#Réseaux de neurones convolutifs(CNN)

Un réseau de neurones à convolution comporte plusieurs couches cachées qui aident à extraire des informations d'une image.

 Les quatres couches importantes de CNN sont :
 1. Convolution layer
 2. ReLU layer
 3. Pooling layer
 4. Fully connected layer

#Exemple d'architecture CNN

Capture_Architecture_cnn.PNG

**Couche convolutive (CONV)**: Ils sont à la base de CNN et sont chargés d'exécuter les opérations de convolution.

Le noyau/filtre est le composant de cette couche qui effectue l'opération de convolution (matrice).

Jusqu'à ce que l'image complète soit numérisée, le noyau effectue des ajustements horizontaux et verticaux en fonction de la vitesse de foulée.

Le noyau est moins volumineux qu'une image, mais il a plus de profondeur.

Cela signifie que si l'image a trois canaux (RGB), la hauteur et la largeur du noyau seront modestes dans l'espace, mais la profondeur couvrira les trois.

Outre la convolution, il existe une autre partie importante des couches convolutives, connue sous le nom de fonction d'activation non linéaire.

Les sorties des opérations linéaires telles que la convolution passent par une fonction d'activation non linéaire.

L'unité linéaire rectifiée (ReLU) est désormais la fonction d'activation non linéaire la plus couramment utilisée. f(x) = max(0, x)

Capture_convnet.PNG

**Pooling Layer (POOL)** : Cette couche est chargée de réduire la dimensionnalité.

Il aide à réduire la quantité de puissance de calcul requise pour traiter les données.


Capture_maxpool.PNG

**Couche entièrement connectée (FC)** : la couche entièrement connectée (FC) fonctionne avec une entrée aplatie, ce qui signifie que chaque entrée est couplée à chaque neurone. Après cela, le vecteur aplati est envoyé via quelques couches FC supplémentaires, où les opérations fonctionnelles mathématiques sont normalement effectuées. La procédure de classification commence à ce stade. Les couches FC se trouvent fréquemment vers la fin des architectures CNN si elles sont présentes.

Capture_fulyconnected.PNG
